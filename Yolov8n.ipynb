```python
from google.colab import drive
import os
from ultralytics import YOLO
import torch
```

```python
dataset_path = '/content/drive/MyDrive/final_dataset_yolo/dataset.yaml'
project_path = '/content/drive/MyDrive/yolov8n_final'
```

```python
# Create project folder if it doesn't exist
os.makedirs(project_path, exist_ok=True)
print(f"Project folder created at: {project_path}")
```

```python
model = YOLO('yolov8n.pt')

model.train(
    data=dataset_path,
    epochs=150,
    imgsz=640,
    batch=16,
    device=0,
    project=project_path,
    name='road_damage_run_n',
    exist_ok=True,
    save_period=5,
    val=True,
    patience=50
)
```

```python
from ultralytics import YOLO


model = YOLO('/content/drive/MyDrive/yolov8n_final/road_damage_run_n/weights/best.pt')


metrics = model.val(
    data='/content/drive/MyDrive/final_dataset_yolo/dataset.yaml',
    split='test',
    conf=0.25,
    iou=0.6,
    plots=True
)


print(metrics)
```

```python
print("=== Accuracy Metrics ===")
print(f"mAP@0.5:       {metrics.box.map50:.4f}")
print(f"mAP@0.5:0.95:  {metrics.box.map:.4f}")
print(f"Precision:     {metrics.box.p.mean():.4f}")
print(f"Recall:        {metrics.box.r.mean():.4f}")
f1 = 2 * (metrics.box.p.mean() * metrics.box.r.mean()) / (metrics.box.p.mean() + metrics.box.r.mean() + 1e-6)
print(f"F1 Score:      {f1:.4f}")
print(f"IoU (mean):    {metrics.box.maps.mean():.4f}")
```

```python
import glob


test_images = glob.glob("/content/drive/MyDrive/final_dataset_yolo/images/test/*.jpg")
sample_img = test_images[0]
print(f"Using test image: {sample_img}")

import time

for _ in range(5):
    _ = model.predict(sample_img, imgsz=640, device=0, verbose=False)

t0 = time.time()
runs = 50
for _ in range(runs):
    _ = model.predict(sample_img, imgsz=640, device=0, verbose=False)
t1 = time.time()

avg_time = (t1 - t0) / runs
fps = 1 / avg_time

print("\n=== Real-Time Performance ===")
print(f"Average Inference Time: {avg_time*1000:.2f} ms")
print(f"FPS: {fps:.2f}")

```
